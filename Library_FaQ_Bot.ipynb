```python
import google.generativeai as genai
```

```python
faq_data = """
Library FAQs:

1. Library timings:
The library is open from 9 AM to 6 PM.

2. Borrow books:
Students can borrow up to 3 books.

3. Late fee:
Late fee is ₹5 per day.

4. WiFi:
Free WiFi is available in the library.

5. Renew books:
Books can be renewed online using the library portal.
"""

from google.colab import userdata
genai.configure(
    api_key=userdata.get("Gemini_API_Key")
)
model= genai.GenerativeModel(
    model_name="gemini-2.5-flash",
    system_instruction=""" You are a helpful Library FAQ chatbot.

    Use the FAQ data whenever the user's question
    is related to library information.

    If the question is general and not related
    to the FAQ data, answer normally using your own knowledge.

    Keep responses clear and friendly.""",
)
chat=model.start_chat(history=[])
```

```python
print("Library FAQ Chatbot ready!")
print("Type 'exit' to stop.\n")

while True:

    user_question = input("User: ")

    if user_question.lower() == "exit":
        print("Chatbot Ended.")
        break

    full_prompt = f"""
    FAQ DATA:
    {faq_data}

    USER QUESTION:
    {user_question}
    """

    response = chat.send_message(
        full_prompt,
        generation_config={
            "temperature": 0.7,
        }
    )

    print("\nBot:", response.text)
    print()
```

```python
print("========")
for i, msg in enumerate(chat.history):
  role = "User" if msg.role == "user" else " BOT"
  print(f"{i+1}. {role}: {msg.parts[0].text}")
```
